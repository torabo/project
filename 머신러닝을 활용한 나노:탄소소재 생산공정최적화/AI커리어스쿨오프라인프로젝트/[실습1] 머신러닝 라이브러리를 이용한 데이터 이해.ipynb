### 1.4 numpy array 생성을 위한 메소드들

**[TODO] 주석에서 요구하는대로 ____을 적절히 수정해보세요.**

np.zeros( *shape* ) $\to$ *shape* 의 모양을 가진 0-배열을 생성합니다.<br>
np.ones( *shape* ) $\to$ *shape* 의 모양을 가진 1-배열을 생성합니다.<br>
np.full( *shape*, *number* ) $\to$ *shape* 의 모양을 가진 *number*-배열을 생성합니다.<br>
np.eye( *number* ) $\to$ (*number*, *number*) 의 모양을 가진 대각행렬을 생성합니다.<br>
np.arange( *number* ) $\to$ *0*~*number-1* 까지의 값을 가지는 배열을 생성합니다.

```python
a = np.zeros((3, 3)) # 모든 원소가 0인 배열
b = np.ones((3, 3)) # 모든 원소가 1인 배열
c = np.full((3, 3), 100) # 모든 원소가 특정 값인 배열
d = np.eye(3) # 크기가 3인 대각행렬
e = np.arange(9).reshape(3, 3) # 0~8 까지로 채워진 배열

# 모든 원소가 0인 배열을 생성합니다.
print('np.zeros((3, 3))')
print(a)
print()

# 모든 원소가 1인 배열을 생성합니다.
print('np.ones((3, 3))')
print(b)
print()

# 모든 원소가 특정 값인 배열을 생성합니다.
print('np.full((3, 3), 100)')
print(c)
print()

# 크기가 3인 대각행렬을 생성합니다.
print('np.eye(3)')
print(d)
print()

# 0~8 까지로 채워진 배열을 생성합니다.
print('np.arange(9).reshape(3, 3)')
print(e)
print()
```

### 3.4 금속분말 데이터 시각화 하기

이번에는, 금속분말 데이터를 시각화해봅시다.

```python
x = data['time_stamp'][:10000:100]
y = data[columns[1]][:10000:100]

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example')
plt.show()
```

X축의 시간이 알아보기 어려우므로, datetime 라이브러리를 이용하여 처리해줍니다.

```python
import datetime
```

```python
times = []
for time_stamp in data['time_stamp']:
    ymd, hhmmss = time_stamp.split(' ')
    y, m, d = ymd.split('-')
    hh, mm, ss = hhmmss.split(':')
    times.append(datetime.datetime(int(y), int(m), int(d), int(hh), int(mm), int(ss)))
```

```python
time_stamp.split(' ')
```

```python
x = times[:10000:100]
y = data[columns[1]][:10000:100]

plt.plot(x, y)
plt.xlabel('X')
plt.xticks(rotation=45)
plt.ylabel('Y')
plt.title('Example')
plt.show()
```

이번에는 여러 칼럼에 대한 그래프를 그려봅니다.

```python
x = times[:10000:100]
y1 = data[columns[1]][:10000:100]
y2 = data[columns[2]][:10000:100]

plt.plot(x, y1, label = columns[1])
plt.plot(x, y2, label = columns[2])
plt.xlabel('X')
plt.xticks(rotation=45)
plt.ylabel('Y')
plt.title('Example')
plt.legend()
plt.show()
```

다음으로, 모든 칼럼에 대한 그래프를 그려봅니다.

```python
x = times[:10000:100]
for column in columns:
    y = data[column][:10000:100]
    plt.plot(x, y, label = column)
plt.xlabel('X')
plt.xticks(rotation=45)
plt.ylabel('Y')
plt.title('Example')
plt.legend()
plt.show()
```

칼럼 수가 너무 많아 알아보기 어렵습니다. 칼럼을 규칙에 따라 나눠봅니다.

```python
columns
```

```python
groups = list(set([group.split('.')[0] for group in columns ]))
groups.sort()
print(groups)
```

```python
columns_group = {}
for group in groups:
    columns_group[group] = []
    i = 0
    for column in columns:
        if group in column:
            columns_group[group].append(column)
            i+=1
    print(group, i)
```

```python
columns_group
```

각 그룹 별로 선그래프를 그려봅니다.

```python
x = times[:10000:100]
for group in columns_group:
    plt.figure(figsize=(20,10))
    columns_ = columns_group[group]
    for column in columns_:
        y = data[column][:10000:100]
        plt.plot(x, y, label = column)
    plt.xlabel('X')
    plt.xticks(rotation=45)
    plt.ylabel('Y')
    plt.title(group)
    plt.legend()
    plt.show()
```
